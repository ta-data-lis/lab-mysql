```python
#DIY

def airquality():
    url = "https://api.tfl.gov.uk/AirQuality"
    response = requests.get(url)
    weather_data = pd.DataFrame(response.json())
    mydict = weather_data["currentForecast"][0]

    return pd.DataFrame.from_dict(mydict, orient="index").T
```

```python
def get_my_tickers(tickers):
    tickers_df = []
    key = "Q3E13KT1Z6IY5Y69"
    url = "https://www.alphavantage.co/query"
    
    for tick in tickers:
        params = {"function":"TIME_SERIES_DAILY_ADJUSTED",
                  "symbol":tick,
                  "apikey":key}
        response = requests.get(url, params = params)

        ticker_daily = pd.DataFrame(response.json()["Time Series (Daily)"]).T
        ticker_daily["Ticker"] = response.json()["Meta Data"]["2. Symbol"]

        tickers_df.append(ticker_daily)
    return tickers_df
```

```python
msft, tsla = get_my_tickers(["MSFT","TSLA"])
```
